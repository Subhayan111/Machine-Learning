```python
import numpy as np
import matplotlib.pyplot as plt

def linear_regression(X, y):
    mean_x = np.mean(X)
    mean_y = np.mean(y)
    diff_x = X - mean_x
    diff_y = y - mean_y
    m = np.sum(diff_x * diff_y) / np.sum(diff_x**2)
    b = mean_y - m * mean_x
    return m, b

def plot_linear_regression(X, y, m, b):
    plt.scatter(X, y, color='blue', label='Data Points')
    plt.plot(X, m * X + b, color='red', label='Linear Regression')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()
    plt.show()

def main():
    # User input for data points
    X_str = input("Enter X values separated by spaces: ")
    y_str = input("Enter corresponding y values separated by spaces: ")

    # Convert input strings to arrays
    X = np.array([float(x) for x in X_str.split()])
    y = np.array([float(y_val) for y_val in y_str.split()])

    # Perform linear regression
    slope, intercept = linear_regression(X, y)

    # Print the equation of the line
    print(f"Linear Regression Equation: y = {slope:.2f} * X + {intercept:.2f}")

    # Plot the linear regression line
    plot_linear_regression(X, y, slope, intercept)

if __name__ == "__main__":
    main()
```
